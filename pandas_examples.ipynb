```python
# source: https://www.kaggle.com/ludobenistant/hr-analytics
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
df = pd.read_csv("HR_comma_sep.csv")
df.head()
```

```python
df.describe()
```

```python
print("The mean satification level was", df["satisfaction_level"].mean())
print("The median satification level was", df["satisfaction_level"].median())
print("The standard deviation of the satification level was", df["satisfaction_level"].std())
print("The variance of the satification level was", df["satisfaction_level"].var())
print("The skew of the satification level was", df["satisfaction_level"].skew())
print("The kurtosis of the satification level was", df["satisfaction_level"].kurt())
```

```python
df["satisfaction_level"].plot.hist()
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, alpha=0.2, figsize=(12, 12), diagonal='kde')
plt.show()
```

```python
from pandas.plotting import andrews_curves

cols = df.columns.tolist()
cols.remove("sales")
no_sales = df[cols]
andrews_curves(no_sales, 'salary')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

parallel_coordinates(no_sales, 'salary')
```

```python
from pandas.plotting import radviz

radviz(no_sales, 'salary')
```
